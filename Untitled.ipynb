```python
car_crashes = sns.load_dataset("car_crashes")
plt.figure(figsize=(10,6))
plt.plot(car_crashes['abbrev'], car_crashes['total'], marker='o', linestyle='-', color='b')
plt.title('Total Car Crashes by State', fontsize=16)
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Crashes', fontsize=12)
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
speeding_avg = car_crashes['speeding'].mean()
non_speeding_avg = 100 - speeding_avg

labels = ['Speeding', 'Non-Speeding']
sizes = [speeding_avg, non_speeding_avg]
colors = ['red', 'green']

plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels, colors=colors, startangle=90, explode=(0.1, 0))
plt.title('Proportion of Crashes Involving Speeding', fontsize=16)
plt.show()
```

```python
alcohol_avg = car_crashes['alcohol'].mean()
non_alcohol_avg = 100 - alcohol_avg

labels = ['Alcohol Involved', 'No Alcohol Involved']
sizes = [alcohol_avg, non_alcohol_avg]
colors = ['orange', 'blue']

plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels, colors=colors, startangle=90, explode=(0.1, 0))
plt.title('Proportion of Crashes Involving Alcohol', fontsize=16)
plt.show()
```

```python
def factorial(n):
    if n == 0:  # Base case
        return 1
    else:
        return n * factorial(n - 1)

    
print(factorial(5))
```
